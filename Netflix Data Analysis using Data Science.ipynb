```python
#In which year highest number of TV Shows and Movies were released? Show with bar graph.
#Formatting release date column
df['Date']=pd.to_datetime(df['Release_Date'])
df.head()
```

```python
#using value_counts() function to know which year highest TV Shows and Movies were released.
df['Date'].dt.year.value_counts()
```

```python
#show with bar graph
df['Date'].dt.year.value_counts().plot(kind='bar')
```

```python
#How many movies and shows are in the dataset? Show with the bar graph
df['Category'].value_counts()
```

```python
plt.figure(figsize=[10,6])
sns.countplot(df['Category'])
```

```python
#Show all the Movies that were released in 2020
df['Year']=df['Date'].dt.year
df.head()
df[(df['Category']=='Movie') & (df['Year']==2020)].head()
```

```python
#Show only the titles of all the TV Shows that were released in Pakistan.
df[(df['Category']=='TV Show')&(df['Country']=='Pakistan')]['Title']
```

```python
#Show top 5 directors who gave highest number of TV Shows and Movies to Netflix.

df['Director'].value_counts().head()
```

```python
#Show all the records where ‘Category is Movies and Type is Comedies’ or ‘Country is United states.

df[(df['Category']=='Movie') & (df['Type']=='Comedies') | (df['Country']=='United States')].head()
```

```python
#Drop the null values from the dataset with the help of dropna() function

df1=df.dropna()
df1.head()
```
